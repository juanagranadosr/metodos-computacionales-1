# Ejercicio 3.

```python
def GaussianD(x, m, sig):
    return np.exp(-(x-m)**2/(2*sig**2))/(sig*(2*np.pi)**(1/2))
```

```python
# par√°metros dados
N = 1000
mu = 2
sigma = 0.5

x = np.zeros(N)
x[0] = 1.01

for i in range(N-1):
    y = x[i] + np.random.uniform(-1,1)
    # metropolis
    if np.random.rand()<=min(1, GaussianD(y, mu, sigma)/GaussianD(x[i], mu, sigma)):
        x[i+1]=y
    else:
        x[i+1]=x[i]
```

```python
plt.hist(x, density=True, bins=30)
x_ =np.linspace(0, 4, 1000)

plt.plot(x_, GaussianD(x_, mu, sigma),'orange')
plt.xlabel('x')
plt.ylabel('Gaussian PDF')
plt.show()
```
