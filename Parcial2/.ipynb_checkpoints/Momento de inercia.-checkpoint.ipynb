```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
from scipy import integrate
from tqdm import tqdm
```

```python
def f(r,z):
    return r*z
```

```python
R = 1
z1 = 0
z2 = 2
N = int(1e5)
Sample = np.zeros(N)
for i in range(N):
    u = np.random.rand()
    r = R*u**(1/2)
    z = np.random.uniform(z1,z2)
    Sample[i] = f(r,z)
```

```python
Vc = np.pi*R**2*(z2-z1)
```

```python
print(np.average(Sample)*Vc,4*np.pi/3)
```

```python
def CreateSphere(N,R=1):
    
    Points = np.zeros((N,3))
        
    for i in tqdm(range(N)):
        
        phi = np.random.uniform(0,2*np.pi)
        u = np.random.rand()
        r = R*u**(1/3)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        Points[i] = [r*np.cos(phi)*np.sin(theta),r*np.sin(phi)*np.sin(theta),r*np.cos(theta)]
        
    return Points
```

```python
Sphere = CreateSphere(100000)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```

```python
def Ixx(y,z):
    return y**2+z**2

def Iyy(x,z):
    return x**2+z**2

def Izz(x,y):
    return x**2+y**2
```

```python
Samplexx = np.zeros(Sphere.shape[0])
Sampleyy = np.zeros(Sphere.shape[0])
Samplezz = np.zeros(Sphere.shape[0])


for i in range(Sphere.shape[0]):
    Samplexx[i] = Ixx(Sphere[i,1],Sphere[i,2])
    Sampleyy[i] = Iyy(Sphere[i,0],Sphere[i,2])
    Samplezz[i] = Izz(Sphere[i,0],Sphere[i,1])

ixx = np.average(Samplexx)
iyy = np.average(Sampleyy)
izz = np.average(Samplezz)

print(ixx, iyy, izz)
```

```python
def Ixy(x,y):
    return x*y

def Iyz(y,z):
    return x+z

def Ixz(x,z):
    return xz
```

```python
Samplexy = np.zeros(Sphere.shape[0])
Sampleyz = np.zeros(Sphere.shape[0])
Samplexz = np.zeros(Sphere.shape[0])


for i in range(Sphere.shape[0]):
    Samplexy[i] = Ixx(Sphere[i,0],Sphere[i,1])
    Sampleyz[i] = Iyy(Sphere[i,1],Sphere[i,2])
    Samplexz[i] = Izz(Sphere[i,0],Sphere[i,2])

ixy = np.average(Samplexy)
iyz = np.average(Sampleyz)
ixz = np.average(Samplexz)

print(ixy, iyz, ixz)
```

Se puede observar que los momentos de inercia son iguales sin importar los ejes de referencia tomados. Esto sucede debido a que se tiene una densidad de masa uniforme por lo que la resistencia al giro en cualquier dirección es igual, es decir, es invariante ante rotaciones.
